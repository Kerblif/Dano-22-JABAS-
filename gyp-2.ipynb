```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def get_tend_ret(db, down = 0, up = 1000):
    db.retired = pd.to_numeric(db.retired)

    db.loc[db.riearn != 0, 'retired'] = 0
    db = db.sort_values(by=['index_wave'])

    missing_part = db[db.rhltc.isin(['.p', '.m', '.d', '.r'])]
    ids_to_drop = missing_part.hhidpn.unique()

    db = db.drop(db[db.hhidpn.isin(ids_to_drop)].index)
    db.rhltc = pd.to_numeric(db.rhltc)

    ids = list(db.hhidpn.unique())

    data = {
        "p3": [],
        "p2": [],
        "p1": [],
        "a1": [],
        "a2": [],
        "a3": []
    }

    for id in ids:
        person_data = db[db.hhidpn == id]
        ret = list(person_data.retired)
        try:
            age = list(map(int, person_data.ragey_b))
        except:
            continue
        
        f = -1

        for i in range(1, len(ret)):
            if ret[i] > ret[i-1]:
                f = i
                break
        
        if f == -1 or not (len(ret) > f + 2 and ret[f] >= 50 and ret[f + 1] >= 50 and ret[f + 2] >= 50 and f - 3 >= 0):
            continue
        rhltc = list(person_data.rhltc)

        if age[f] >= down and age[f] <= up:
            data["p3"].append(3 - rhltc[f - 3])
            data["p2"].append(3 - rhltc[f - 2])
            data["p1"].append(3 - rhltc[f - 1])
            data["a1"].append(3 - rhltc[f])
            data["a2"].append(3 - rhltc[f + 1])
            data["a3"].append(3 - rhltc[f + 2])
    r = pd.DataFrame(data)

    a = [0]

    c = list(r.columns)

    for i in c:
        t = r[i]
        t = t.sum() / t.count()
        a.append(a[-1] + t)
    
    return a
```

```python
def get_tend_work(db, down):
    db.retired = pd.to_numeric(db.retired)

    db.loc[db.riearn != 0, 'retired'] = 0
    db = db.sort_values(by=['index_wave'])

    missing_part = db[db.rhltc.isin(['.p', '.m', '.d', '.r'])]
    ids_to_drop = missing_part.hhidpn.unique()

    db = db.drop(db[db.hhidpn.isin(ids_to_drop)].index)
    db.rhltc = pd.to_numeric(db.rhltc)

    ids = list(db.hhidpn.unique())

    data = {
        "p3": [],
        "p2": [],
        "p1": [],
        "a1": [],
        "a2": [],
        "a3": []
    }

    for id in ids:
        person_data = db[db.hhidpn == id]
        ret = list(person_data.retired)
        try:
            age = list(map(int, person_data.ragey_b))
        except:
            continue
        

        if not down in age:
            continue
        
        f = age.index(down)
        
        if not (len(ret) > f + 2 and f - 3 >= 0):
            continue
        rhltc = list(person_data.rhltc)

        data["p3"].append(3 - rhltc[f - 3])
        data["p2"].append(3 - rhltc[f - 2])
        data["p1"].append(3 - rhltc[f - 1])
        data["a1"].append(3 - rhltc[f])
        data["a2"].append(3 - rhltc[f + 1])
        data["a3"].append(3 - rhltc[f + 2])
    r = pd.DataFrame(data)

    a = [0]

    c = list(r.columns)

    for i in c:
        t = r[i]
        t = t.sum() / t.count()
        a.append(a[-1] + t)
    
    return a
```
